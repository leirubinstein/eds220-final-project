```python
# Import Phoenix subdivision shapefile
root = os.path.join('/','Users', 'rubinstein', 'MEDS', 'EDS-220-ENV-DATASETS', 'eds220-final-project', 'data')
az = gpd.read_file(os.path.join(root,
                                     'tl_2022_04_cousub',
                                     'tl_2022_04_cousub.shp'))
# Reproject CRS to EPSG:4326
az = az.to_crs('epsg:4326')

# Filter to Phoenix subdivision
phoenix = az[az['NAME'] == 'Phoenix']
```

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.axis('off')

az.plot(ax=ax, color='tan')

phoenix.plot(ax=ax, color='saddlebrown')

ax.legend(fontsize='small')
plt.tight_layout()
plt.subplots_adjust(bottom=0.05)

ax.set_title('Phoenix Subdivision (in brown)')

# plt.figtext(0.01, 0.015, "Data: NASA Landsat & CalFire", ha="left", fontsize=10)

plt.show()
```

```python
# Temporal range of interest during growth of development
time_range = '2017-01-01/2021-01-01'
```

```python
# Create bounding box for search
bbox = [-112.826843, 32.974108, -111.184387, 33.863574]
```

```python
# Access MPC catalog
catalog = Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1",
    modifier=planetary_computer.sign_inplace,
)

# Search MPC catalog
search = catalog.search(collections = ['io-biodiversity'],
                        bbox = bbox,
                        datetime = time_range
                       )

# Retrieve search items
items = search.item_collection()
print(f"Returned {len(items)} Items")
items
```

```python
# Get the 2017 and 2020 BII items
item_2020 = items[0]
item_2017 = items[3]
```

```python
# Access raster data from item
io_bio2020 = rioxr.open_rasterio(item_2020.assets['data'].href)
io_bio2017 = rioxr.open_rasterio(item_2017.assets['data'].href)
```

```python
io_bio2020 = (io_bio2020.rio.clip_box(*phoenix.total_bounds)
                 .rio.clip(phoenix.geometry))
io_bio2020 = io_bio2020.squeeze().drop_vars('band')

io_bio2017 = (io_bio2017.rio.clip_box(*phoenix.total_bounds)
                 .rio.clip(phoenix.geometry))
io_bio2017 = io_bio2017.squeeze().drop_vars('band')
```

### Calculate the percentage of area of the Phoenix subdivision with a BII of at least 0.75 in 2017 & 2020

```python
print("2020 Min value:", io_bio2020.min().values)
print("2020 Max value:", io_bio2020.max().values)
print("2020 Number of valid pixels:", io_bio2020.count().values)
```

```python
print("2017 Min value:", io_bio2017.min().values)
print("2017 Max value:", io_bio2017.max().values)
print("2017 Number of valid pixels:", io_bio2017.count().values)
```

```python
high_bii_2020 = (io_bio2020 >= 0.75)
percentage_2020 = (high_bii_2020.sum() / io_bio2020.count()) * 100

high_bii_2017 = (io_bio2017 >= 0.75)
percentage_2017 = (high_bii_2017.sum() / io_bio2017.count()) * 100

print(f"2017 Percentage of area of the Phoenix subdivision with BII >= 0.75: {percentage_2017.values:.2f}%")
print(f"2020 Percentage of area of the Phoenix subdivision with BII >= 0.75: {percentage_2020.values:.2f}%")
```

```python
lost_bii = high_bii_2017 ^ high_bii_2020
type(lost_bii)
```

```python
lost_bii2 = xr.where(lost_bii, 1, np.nan)
# lost_bii2 = xr.where(~np.isnan(lost_bii2), 1, np.nan)
lost_bii2
```

```python
aspect = io_bio2020.rio.width/io_bio2020.rio.height
fig, ax = plt.subplots(figsize=(12, 12/aspect))

# First plot
im1 = io_bio2020.plot(
    ax=ax,
    cmap='Greens',
    add_labels=False,
    add_colorbar=True,
    cbar_kwargs={
        'label': 'Biodiversity Intactness Index',
        'fraction': 0.046,   
        'aspect': 30,
    },
)

# Second plot with a single-color colormap
im2 = lost_bii2.plot(
    ax=ax,
    cmap=plt.cm.colors.ListedColormap(['red']),
    add_labels=False,
    add_colorbar=False,
)

# Add a simple legend for the binary layer
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='red', label='Areas of BII Loss 2017-2020')]
ax.legend(handles=legend_elements, loc='upper left')

ax.axis('off')

# Add title
plt.title('Biodiversity Intactness Index Changes in Phoenix', pad=20, size=14)
```

```python
aspect=io_bio2020.rio.width/io_bio2020.rio.height
fig, ax = plt.subplots(figsize=(12, 12/aspect))

# First plot
im1 = io_bio2020.plot(ax=ax,
                      cmap='Greens',
                      add_labels=False,
                      add_colorbar=True)

# Second plot with a single-color colormap
im2 = lost_bii2.plot(ax=ax,
                     cmap=plt.cm.colors.ListedColormap(['red']),  
                     add_labels=False,
                     add_colorbar=False)

# Add a simple legend for the binary layer
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='red', label='Areas of BII Loss 2017-2020')]
ax.legend(handles=legend_elements, loc='upper left')

ax.axis('off')

# Add title
plt.title('Biodiversity Intactness Index Changes in Phoenix', pad=20, size=14)
```

```python
# Plot raster with correct ratio
aspect=io_bio2020.rio.width/io_bio2020.rio.height
fig, ax = plt.subplots(figsize=(12, 12/aspect))

io_bio2020.plot(ax=ax,
                cmap='Greens',
                linewidth=1,
                add_labels=False,
                add_colorbar=True)

lost_bii.plot(ax=ax,
               cmap='Reds',
               add_labels=True,
               add_colorbar=False,
               label='Areas of BII Loss 2017-2020'
              )

ax.axis('off')


# Add title
plt.title('Biodiversity Intactness Index Changes in Phoenix', pad=20, size=14)
```
